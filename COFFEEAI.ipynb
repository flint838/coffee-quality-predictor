```python
from pyspark.sql import SparkSession
df = SparkSession.builder.appName('coffee').getOrCreate()
sparkdf = df.read.csv('/home/salahuddin/Downloads/coffeeData/df_arabica_clean.csv',header=True,inferSchema=True)
sparkdf.show() 
```

```python
#here we got the column names and types in array form to be used whenever needed
columns = []
columns_type=[]
i = 0
for col in sparkdf.schema.fields:
    if i==0:
        i+=1
        continue
    columns.append(col.name)
    columns_type.append(str(col.dataType))
columns_type
```

```python
#array of names of columns to drop
to_drop = []
for i,j in zip(columns,columns_type):
    if j == 'StringType()':
        to_drop.append(i)
to_drop        
```

```python
newdf=sparkdf.drop(*to_drop)
newdf.show()
```

```python
#further processing
to_drop = ['ID','Number of Bags','Body','Overall','Total Cup points','Moisture Percentage','Quakers','Defects']
newdf = newdf.drop(*to_drop)
newdf.show()
```

```python
import pandas as pd
newpandasdf = newdf.to_pandas_on_spark()
newpandasdf.head(4)
```

```python
newpandasdf["overall_coffee_points"]  = newpandasdf['Aroma']+newpandasdf['Flavor']+newpandasdf['Aftertaste']+newpandasdf['Acidity']+newpandasdf['Balance']+newpandasdf['Uniformity']+newpandasdf['Clean Cup']+newpandasdf['Sweetness']-((4*newpandasdf['Category One Defects'])+(4*newpandasdf['Category Two Defects']))
```

```python
newpandasdf
```

```python
#if it is above 65--> very good//between 55 and 65--> good//between 45 and 55 average// below 45-->not accepted
import numpy as np
counts = newpandasdf['overall_coffee_points']
counts.to_numpy()
counts
```

```python
quality = []
for i in range(len(counts)):
    if counts[i] > 65.00:
        quality.append(4)
    elif counts[i] > 55.00 and counts[i] <= 65.00:
        quality.append(3)
    elif counts[i] > 45.00 and counts[i] <= 55.00:
        quality.append(2)
    elif counts[i] <= 45.00:
        quality.append(0)
    else:
        quality.append(-1) 
quality = pd.DataFrame(quality)
               

    
```

```python
quality.columns = ['quality']
quality
```

```python
qual =  quality['quality'].to_list()
newpandasdf['quality'] = qual 
newpandasdf
```
